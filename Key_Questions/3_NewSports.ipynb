### Import libraries and read the data

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np


df = pd.read_csv("clean_olympics_dataset.csv")
```

### Check the data imported properly

```python
df.head(10)
```

### Which year did each sport make its first appearance?
Table as an output

```python
t = df.groupby('Sport')['Year'].min().reset_index().sort_values('Year', ascending = False)
```

```python
t.style.hide(axis='index')
```

### Which year did a team GB athlete make their first appearance for each sport?
Table as an output (add a column to the prior table)

```python
# Find the first year for each sport overall
first_year_overall = df.groupby('Sport')['Year'].min().reset_index()
first_year_overall.columns = ['Sport', 'First_Year']

# Find the first year for each sport for Great Britain
first_year_gb = df[df['Country'] == 'GREAT BRITAIN'].groupby('Sport')['Year'].min().reset_index()
first_year_gb.columns = ['Sport', 'First_Year_GB']

# Merge the two DataFrames
t = pd.merge(first_year_overall, first_year_gb, on='Sport', how='inner')

# Select only the required columns
t = t[['Sport', 'First_Year', 'First_Year_GB']].sort_values(by='First_Year_GB')

# Display the result
t.style.hide(axis='index')
```

### Which sports did team GB not send athletes in the first year the sport appeared at the Olympics?

```python
# Create a new DataFrame with rows where First_Year and First_Year_GB are different
different_years = t[t['First_Year'] != t['First_Year_GB']]

# Sort the result by First_Year_GB
different_years = different_years.sort_values(by='First_Year_GB')

# Display the result
different_years.style.hide(axis='index')
```

### Which sports have been removed from the Olympics?
Table shows last year appeared at the Olympics

```python
t = df.groupby('Sport')['Year'].max().reset_index().sort_values('Year', ascending = True)
```
